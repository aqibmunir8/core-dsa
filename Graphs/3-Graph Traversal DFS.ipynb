### Graph Traversal DFS

```python
class Graph:
    def __init__(self, size):
        self.size = size
        # Standard structural initialization: creating the square matrix grid
        self.matrix = [[0] * size for _ in range(size)]

    def add_edge(self, source, destination):
        # Adding structural undirected connection pathways
        self.matrix[source][destination] = 1
        self.matrix[destination][source] = 1

    def dfs(self, source):
        # 1. Initialize the Visited List tracking mechanism
        visited = [False] * self.size
        
        # 2. Set up the control Stack structure and seed it with the source node
        stack = []
        stack.append(source)
        
        # 3. Process data sequentially until structural elements clear out
        while len(stack) > 0:
            # Retrieve the most recent targeted item
            v = stack.pop()
            
            # Execute processing only if the node has not been captured yet
            if not visited[v]:
                print(v, end=" ")
                visited[v] = True  # Flag node as officially tracked
                
                # Scan structural adjacency matrix row to identify neighbors
                for i in range(self.size):
                    # If an edge exists and the neighboring node is unvisited, push it
                    if self.matrix[v][i] == 1 and not visited[i]:
                        stack.append(i)
```

```python
g = Graph(6)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(2, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)
g.add_edge(4, 5)

# Execute DFS traversal starting from Vertex index 0
print("DFS Traversal Output:")
g.dfs(0)
```
