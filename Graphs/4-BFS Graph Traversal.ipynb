#### BFS Graph Traversal

```python
from collections import deque

class Graph:
    def __init__(self, size):
        self.size = size
        # Standard structural initialization: creating the square matrix grid
        self.mat = [[0] * size for x in range(size)]

    def add_edge(self, source, destination):
        if(0 <= source < self.size and 0<= destination < self.size):
            self.mat[source][destination] = 1
            self.mat[destination][source] = 1
        else: 
            print("Invalid Edge")
    
    def bfs(self, src):
        visited = [False] * self.size
        queue = deque([src])
        visited[src] = True

        while queue:
            v = queue.popleft()
            print(v, end=" ")

            for i in range(self.size):
                if(self.mat[v][i] == 1 and visited[i] == False):
                    visited[i] = True
                    queue.append(i)

   
```

```python
# Instantiate an 8-node structural graph space
g = Graph(8)

# Configure link edges
g.add_edge(0, 1)
g.add_edge(0, 3)
g.add_edge(1, 3)
g.add_edge(3, 5)
g.add_edge(3, 4)
g.add_edge(4, 5)
g.add_edge(4, 6)
g.add_edge(6, 2)
g.add_edge(6, 7)

# Run Level Order BFS Traversal starting from vertex index 0
print("BFS Traversal Output:")
g.bfs(0)
```
