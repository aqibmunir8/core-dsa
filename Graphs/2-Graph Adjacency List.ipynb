```python
class Graph:
    def __init__(self):
        # Empty dictionary to capture dynamic keys and structural values
        self.adj_list = {}

    def add_vertex(self, vertex):
        # Ensure duplicate vertex keys are never overwriting data
        if vertex not in self.adj_list:
            self.adj_list[vertex] = []

    def add_edge(self, source, destination):
        # Dynamically verify / add vertices into the workspace
        self.add_vertex(source)
        self.add_vertex(destination)

        # Connect Source to Destination
        self.adj_list[source].append(destination)

        # For Undirected Graphs, cross-link Destination back to Source
        self.adj_list[destination].append(source)
            
    def print_graph(self):
        for vertex in self.adj_list:
            print(f"{vertex} -> {self.adj_list[vertex]}")
```

```python
g = Graph()
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(1, 4)
g.add_edge(4, 3)
g.add_edge(2, 4)
g.add_edge(4, 5)
g.add_edge(3, 5)

g.print_graph()
```
